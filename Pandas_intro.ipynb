```python
#using loc, ilox and ix
#create a dataframe
songs = {
            "Artist": ["jackson","Beatols","Calcutta","CarlBrave"],
            "Album": ["thriller","hey jude","Paracetamolo","stanza 126"],
            "Realesed": [1982,1980,2018,2019],
            "lenght":["2h","1.5h","2h","2h"]
          }

df2 = pd.DataFrame(songs)

print (df2)
```

```python
#how to select data with loc
#loc is primarily label based; when two arguments are used, you use column headers and row indexes to select the data you want.
#loc can also take an integer as a row or column number.
df2.loc[0,"Album"]#thriller
```

```python
df2.loc[3,"lenght"]#2h
```

```python
#iloc is integer-based. You use column numbers and row numbers to get rows or columns at particular positions in the data frame.#
df2.iloc[0,0]
```

```python
df2.iloc[0,2]
```

```python
#using loc and iloc for slicing#
```

```python
z=df2.loc[0:2,"Album":"Artist"]
print(z)
```

```python
k=df2.iloc[0:2,0:3]
print(k)
```

```python
#Pandas has the method unique to determine the unique elements in a column of a data frame. 
#Lets say we would like to determine the unique year of the albums in the data set.#
df2["Realesed"].unique
```

```python
#Let's say we would like to create a new database consisting of songs from the 2000s and after.
df2["Realesed"]>=2000
```

```python
#we can create a new data frame with song from the 2000s and after using:
df3=df[df2["Realesed"]>=2000]
print(df3)
```

```python
#We can save the new data frame 
df3.to_csv("nuovo.csv")
```
